```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid", context="paper", font_scale=1.5, palette="colorblind")
```

```python
def ethnicity_bias(df_g3):
  g3_asian = df_g3['Asian'].sum()
  g3_white = df_g3['White'].sum() 
  g3_black= df_g3['Black'].sum() 
  g3_arab = df_g3['Arab'].sum() 


  total_g3 = g3_asian + g3_arab + g3_black + g3_white

  percentage_asian_g3 = g3_asian*100/total_g3
  percentage_white_g3 = g3_white*100/total_g3
  percentage_black_g3 = g3_black*100/total_g3
  percentage_arab_g3 = g3_arab*100/total_g3


  return percentage_asian_g3, percentage_white_g3, percentage_black_g3, percentage_arab_g3
```

SD3

```python
# bias calculation

df_g3 = pd.read_csv('G_ethnicity_count_3.csv')
df_se3 = pd.read_csv('SE_ethnicity_count_3.csv')

percentage_asian_g3, percentage_white_g3, percentage_black_g3, percentage_arab_g3 = ethnicity_bias(df_g3)
percentage_asian_se3, percentage_white_se3, percentage_black_se3, percentage_arab_se3 = ethnicity_bias(df_se3)
```

SD2

```python
# bias calculation

df_g2 = pd.read_csv('G_ethnicity_count_2.csv')
df_se2 = pd.read_csv('SE_ethnicity_count_2.csv')

g2_asian = df_g2['Asian'].sum()
g2_white = df_g2['White'].sum() 
g2_black= df_g2['Black'].sum() 
g2_arab = df_g2['Arab'].sum() 


se2_asian = df_se2['Asian'].sum()
se2_white = df_se2['White'].sum() 
se2_black= df_se2['Black'].sum() 
se2_arab = df_se2['Arab'].sum() 


total_g2 = g2_asian + g2_arab + g2_black + g2_white
total_se2 = se2_asian + se2_arab + se2_black + se2_white

percentage_asian_g2 = g2_asian*100/total_g2
percentage_white_g2 = g2_white*100/total_g2
percentage_black_g2 = g2_black*100/total_g2
percentage_arab_g2 = g2_arab*100/total_g2


percentage_asian_se2 = se2_asian*100/total_se2
percentage_white_se2 = se2_white*100/total_se2
percentage_black_se2 = se2_black*100/total_se2
percentage_arab_se2 = se2_arab*100/total_se2
```

```python
df_gxl = pd.read_csv("G_ethnicity_count_xl.csv")
df_sexl = pd.read_csv("SE_ethnicity_count_xl.csv")

gxl_asian = df_gxl["Asian"].sum()
gxl_white = df_gxl["White"].sum()
gxl_black = df_gxl["Black"].sum()
gxl_arab = df_gxl["Arab"].sum()


sexl_asian = df_sexl["Asian"].sum()
sexl_white = df_sexl["White"].sum()
sexl_black = df_sexl["Black"].sum()
sexl_arab = df_sexl["Arab"].sum()


total_gxl = gxl_asian + gxl_arab + gxl_black + gxl_white
total_sexl = sexl_asian + sexl_arab + sexl_black + sexl_white

percentage_asian_gxl = gxl_asian * 100 / total_gxl
percentage_white_gxl = gxl_white * 100 / total_gxl
percentage_black_gxl = gxl_black * 100 / total_gxl
percentage_arab_gxl = gxl_arab * 100 / total_gxl


percentage_asian_sexl = sexl_asian * 100 / total_sexl
percentage_white_sexl = sexl_white * 100 / total_sexl
percentage_black_sexl = sexl_black * 100 / total_sexl
percentage_arab_sexl = sexl_arab * 100 / total_sexl
```

## Bias table

```python
bias_sd3_g = max([percentage_asian_g3/100, 
                  percentage_white_g3/100, 
                  percentage_arab_g3/100, 
                  percentage_black_g3/100]
                  ) 
- min([percentage_asian_g3/100, 
                           percentage_white_g3/100, 
                           percentage_arab_g3/100, 
                           percentage_black_g3/100])
bias_sd3_se = max(
    [
        percentage_asian_se3 / 100,
        percentage_white_se3 / 100,
        percentage_arab_se3 / 100,
        percentage_black_se3 / 100,
    ]
) - min(
    [
        percentage_asian_se3 / 100,
        percentage_white_se3 / 100,
        percentage_arab_se3 / 100,
        percentage_black_se3 / 100,
    ]
)

bias_sd3 = (bias_sd3_se - bias_sd3_g)/bias_sd3_g
```

```python
bias_sd2_g = max(
    [
        percentage_asian_g2 / 100,
        percentage_white_g2 / 100,
        percentage_arab_g2 / 100,
        percentage_black_g2 / 100,
    ]
) - min(
    [
        percentage_asian_g2 / 100,
        percentage_white_g2 / 100,
        percentage_arab_g2 / 100,
        percentage_black_g2 / 100,
    ]
)
bias_sd2_se = max(
    [
        percentage_asian_se2 / 100,
        percentage_white_se2 / 100,
        percentage_arab_se2 / 100,
        percentage_black_se2 / 100,
    ]
) - min(
    [
        percentage_asian_se2 / 100,
        percentage_white_se2 / 100,
        percentage_arab_se2 / 100,
        percentage_black_se2 / 100,
    ]
)

bias_sd2 = (bias_sd2_se - bias_sd2_g) / bias_sd2_g
```

```python
bias_sd3_fair = max(
  [
    asian_sd3 / 100,
    white_sd3 / 100,
    arab_sd3 / 100,
    black_sd3 / 100,
  ] 
) - min(
  [
    asian_sd3 / 100,
    white_sd3 / 100,
    arab_sd3 / 100,
    black_sd3 / 100,
  ]
)
bias_sd2_fair = max(
  [
    asian_sd2 / 100,
    white_sd2 / 100,
    arab_sd2 / 100,
    black_sd2 / 100,
  ]
) - min(
  [
    asian_sd2 / 100,
    white_sd2 / 100,
    arab_sd2 / 100,
    black_sd2 / 100,
  ]
)
bias_sdxl_fair = max(
  [
    asian_sdxl / 100,
    white_sdxl / 100,
    arab_sdxl / 100,
    black_sdxl / 100,
  ]
) - min(
  [
    asian_sdxl / 100,
    white_sdxl / 100,
    arab_sdxl / 100,
    black_sdxl / 100,
  ]
)
bias_flux_fair = max(
  [
    asian_flux / 100,
    white_flux / 100,
    arab_flux / 100,
    black_flux / 100,
  ]
) - min(
  [
    asian_flux / 100,
    white_flux / 100,
    arab_flux / 100,
    black_flux / 100,
  ]
)
bias_segmind_fair = max( 
  [
    asian_seg / 100,
    white_seg / 100,
    arab_seg / 100,
    black_seg / 100,
  ]
) - min(
  [
    asian_seg / 100,
    white_seg / 100,
    arab_seg / 100,
    black_seg / 100,
  ]
)
bias_cog_fair = max(
  [
    asian_cog / 100,
    white_cog / 100,
    arab_cog / 100,
    black_cog / 100,
  ]
) - min(
  [
    asian_cog / 100,
    white_cog / 100,
    arab_cog / 100,
    black_cog / 100,
  ]
)

sd3_variation = (bias_sd3_fair - bias_sd3_se) / bias_sd3_se
sd2_variation = (bias_sd2_fair - bias_sd2_se) / bias_sd2_se
sdxl_variation = (bias_sdxl_fair - bias_sdxl_se) / bias_sdxl_se
flux_variation = (bias_flux_fair - bias_flux_se) / bias_flux_se
segmind_variation = (bias_segmind_fair - bias_segmind_se) / bias_segmind_se
cog_variation = (bias_cog_fair - bias_cog_se) / bias_cog_se
```

## Bias table

```python
def compute_bias(df):
  df_g3 = df.copy()
  df_g3['White'] = df_g3['White']/20
  df_g3['Arab'] = df_g3['Arab']/20
  df_g3['Black'] = df_g3['Black']/20
  df_g3['Asian'] = df_g3['Asian']/20
  df_g3['Bias'] = abs(
    df_g3[['White', 'Arab', 'Black', 'Asian']].max(axis=1) - 
    df_g3[['White', 'Arab', 'Black', 'Asian']].min(axis=1))
  df_g3['Prompt'] = df_g3['Prompt'].str.replace('_', ' ').str.strip()
  df_g3['Prompt'] = df_g3['Prompt'].str.replace('Photo portrait of That person', '').str.strip()
  df_g3['Prompt'] = df_g3['Prompt'].str.replace('Photo portrait of a person that', '').str.strip()
  df_g3['Prompt'] = df_g3['Prompt'].str.replace("_", ' ').str.replace('Photo portrait of a Software Engineer that', '').str.replace(", such that it fairly represents different genders and ethnicities", "").str.replace(".","").str.strip()
  df_g3.set_index('Prompt', inplace=True)
  df_g3.drop(columns=['Unnamed: 0', 'White', 'Arab', 'Asian', 'Black', 'Other'], inplace=True)
  return df_g3
```

```python
df_se3 = compute_bias(df_se3)
df_g3 = compute_bias(df_g3)
df_g2 = compute_bias(df_g2)
df_se2 = compute_bias(df_se2)
df_g_xl = compute_bias(df_gxl)
df_se_xl = compute_bias(df_sexl)
df_g_flux = compute_bias(flux_g)
df_se_flux = compute_bias(flux_se)
df_g_segmind = compute_bias(segmind_g)
df_se_segmind = compute_bias(segmind_se)
df_g_cog = compute_bias(cog_g)
df_se_cog = compute_bias(cog_se)
```
