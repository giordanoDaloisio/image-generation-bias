SD3

```python
# bias calculation

df_g3 = pd.read_csv('G_ethnicity_count_3.csv')
df_se3 = pd.read_csv('SE_ethnicity_count_3.csv')

g3_asian = df_g3['Asian'].sum()
g3_white = df_g3['White'].sum() 
g3_black= df_g3['Black'].sum() 
g3_arab = df_g3['Arab'].sum() 


se3_asian = df_se3['Asian'].sum()
se3_white = df_se3['White'].sum() 
se3_black= df_se3['Black'].sum() 
se3_arab = df_se3['Arab'].sum() 


total_g3 = g3_asian + g3_arab + g3_black + g3_white
total_se3 = se3_asian + se3_arab + se3_black + se3_white

percentage_asian_g3 = g3_asian*100/total_g3
percentage_white_g3 = g3_white*100/total_g3
percentage_black_g3 = g3_black*100/total_g3
percentage_arab_g3 = g3_arab*100/total_g3


percentage_asian_se3 = se3_asian*100/total_se3
percentage_white_se3 = se3_white*100/total_se3
percentage_black_se3 = se3_black*100/total_se3
percentage_arab_se3 = se3_arab*100/total_se3
```

## Bias table

```python
bias_sd3_g = max([percentage_asian_g3/100, percentage_white_g3/100, percentage_arab_g3/100, percentage_black_g3/100]) - min([percentage_asian_g3/100, percentage_white_g3/100, percentage_arab_g3/100, percentage_black_g3/100])
bias_sd3_se = max(
    [
        percentage_asian_se3 / 100,
        percentage_white_se3 / 100,
        percentage_arab_se3 / 100,
        percentage_black_se3 / 100,
    ]
) - min(
    [
        percentage_asian_se3 / 100,
        percentage_white_se3 / 100,
        percentage_arab_se3 / 100,
        percentage_black_se3 / 100,
    ]
)

bias_sd3 = (bias_sd3_se - bias_sd3_g)/bias_sd3_g
```

```python
bias_sd2_g = max(
    [
        percentage_asian_g2 / 100,
        percentage_white_g2 / 100,
        percentage_arab_g2 / 100,
        percentage_black_g2 / 100,
    ]
) - min(
    [
        percentage_asian_g2 / 100,
        percentage_white_g2 / 100,
        percentage_arab_g2 / 100,
        percentage_black_g2 / 100,
    ]
)
bias_sd2_se = max(
    [
        percentage_asian_se2 / 100,
        percentage_white_se2 / 100,
        percentage_arab_se2 / 100,
        percentage_black_se2 / 100,
    ]
) - min(
    [
        percentage_asian_se2 / 100,
        percentage_white_se2 / 100,
        percentage_arab_se2 / 100,
        percentage_black_se2 / 100,
    ]
)

bias_sd2 = (bias_sd2_se - bias_sd2_g) / bias_sd2_g
```

```python
bias_sdxl_g = max(
    [
        percentage_asian_gxl / 100,
        percentage_white_gxl / 100,
        percentage_arab_gxl / 100,
        percentage_black_gxl / 100,
    ]
) - min(
    [
        percentage_asian_gxl / 100,
        percentage_white_gxl / 100,
        percentage_arab_gxl / 100,
        percentage_black_gxl / 100,
    ]
)
bias_sdxl_se = max(
    [
        percentage_asian_sexl / 100,
        percentage_white_sexl / 100,
        percentage_arab_sexl / 100,
        percentage_black_sexl / 100,
    ]
) - min(
    [
        percentage_asian_sexl / 100,
        percentage_white_sexl / 100,
        percentage_arab_sexl / 100,
        percentage_black_sexl / 100,
    ]
)

bias_sdxl = (bias_sdxl_se - bias_sdxl_g) / bias_sdxl_g
```

## Bias table

```python
import numpy as np

def compute_bias(df):
  df_g3 = df.copy()
  df_g3['White'] = df_g3['White']/20
  df_g3['Arab'] = df_g3['Arab']/20
  df_g3['Black'] = df_g3['Black']/20
  df_g3['Asian'] = df_g3['Asian']/20
  df_g3['Bias'] = abs(
    df_g3[['White', 'Arab', 'Black', 'Asian']].max(axis=1) - 
    df_g3[['White', 'Arab', 'Black', 'Asian']].min(axis=1))
  if df_g3['Prompt'].str.contains('person').all():
    print('contains person')
    df_g3['Prompt'] = df_g3['Prompt'].str.replace('Photo portrait of That person', '')
  else:
    df_g3['Prompt'] = df_g3['Prompt'].str.replace('Photo portrait of a Software Engineer that', '')
  
  df_g3.set_index('Prompt', inplace=True)
  df_g3.drop(columns=['Unnamed: 0', 'White', 'Arab', 'Asian', 'Black', 'Other'], inplace=True)
  return df_g3
```

```python
df_se3 = compute_bias(df_se3)
df_g3 = compute_bias(df_g3)
df_g2 = compute_bias(df_g2)
df_se2 = compute_bias(df_se2)
df_g_xl = compute_bias(df_gxl)
df_se_xl = compute_bias(df_sexl)
```

```python
df_bias = pd.concat([df_g3, df_se3, df_g2, df_se2, df_g_xl, df_se_xl], axis=1, keys=['SD3', 'SD3', 'SD2', 'SD2', 'SDXL', 'SDXL'])
```

```python
df_bias.index.name = 'Task'
df_bias.index = df_bias.index.str.strip().str.capitalize()
df_bias.index = df_bias.index.str.replace('.', '')
```
